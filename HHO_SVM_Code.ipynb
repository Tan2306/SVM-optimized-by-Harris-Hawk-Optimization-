```python
data = pd.read_csv('Placement_Data_Full_Class.csv')
# data.dropna(axis=0,how='any', inplace=True)
# Import LabelEncoder
from sklearn import preprocessing

# Creating labelEncoder
le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
data['gender']=le.fit_transform(data['gender'])
data['ssc_b']=le.fit_transform(data['ssc_b'])
data['hsc_b']=le.fit_transform(data['hsc_b'])
data['hsc_s']=le.fit_transform(data['hsc_s'])
data['degree_t']=le.fit_transform(data['degree_t'])
data['workex']=le.fit_transform(data['workex'])
data['specialisation']=le.fit_transform(data['specialisation'])

# Spliting data into Features and Target
features=data[['gender', 'ssc_p', 'ssc_b', 'hsc_p', 'hsc_b', 'hsc_s',
       'degree_p', 'degree_t', 'workex', 'etest_p', 'specialisation', 'mba_p']]
target=data['status']
```

## Why do we need nature based optimization at all?
### We see that the relationship between C, gamma and the accuracy is non deterministic. It depends on the type of dataset that we are dealing with. The following grapgh showcases the randomness with which these parameters seem to be related with the accuracy. 

```python
def TestFunc(position): 
    cv = KFold(n_splits=10, shuffle=True)
    model=SVC(kernel='rbf',gamma=position[0],C=position[1])
    scores = cross_val_score(model, features, target, scoring='accuracy', cv=cv, n_jobs=-1)
    return(mean(scores))

gamma=[]
C=[]
accuracy=[]

for i in range(1,10):
    for j in range(1,10):
        accuracy.append(TestFunc([i/10,j/10]))
        gamma.append(i)
        C.append(j)
        
fig = plt.figure()
ax = fig.add_subplot(111,projection = "3d")
ax.scatter(xs = gamma, ys = C, zs = accuracy)
ax.set_xlabel("Gamma")
ax.set_ylabel("C")
ax.set_zlabel("Accuracy")
```

```python

def LF():                                           #Levy Flight
    u = random.random()
    v = random.random()
    beta = 1.5
    sigma = math.gamma(1+beta)*math.sin(math.pi*beta*0.5)
    sigma = sigma/(math.gamma(1+beta*0.5)*beta*math.pow(2,(beta-1)/2))
    sigma = math.pow(sigma,1/beta) 
    result = 0.01 * u * sigma / math.pow(abs(v),1/beta)
    return result


def BringRange(position):                          #Function responsible for bringing X[i] into range
    ubcond=position>ub
    lbcond=position<lb
    for j in range(dim):
        if ubcond[j]==True:
            position[j]=ub[j]
        if lbcond[j]==True:
            position[j]=lb[j]
            

def FitnessFunc(position):                         #Fitness Function
    cv = KFold(n_splits=10, shuffle=True)
    model=SVC(kernel='rbf',gamma=position[0],C=position[1])
    scores = cross_val_score(model, features, target, scoring='accuracy', cv=cv, n_jobs=-1)
    return(mean(scores))


result=[]



for temp in range(30):
    print()
    print()
    print("temp=",temp)               #Initializing
    N = 30                            #Number of search agents (ideally 30)
    dim = 2                           #Number of parameters we are trying to optimize (C and gamma for SVM)
    lb=np.array([0.001,0.001])        #The lower bound of C and gamma
    ub=np.array([1000,100])           #The upper bound of C and gamma
    T = 20                            #Number of iterations (ideally 100)

    X = np.random.rand(N,dim)         #The initial population (random)


    for i in range(N):                #Checking the Values - To make sure they are in range
        X[i]=lb+X[i]*(ub-lb) 


    fitness=[]                                  #Calculating Fitness for each Hawk  
    for i in range(N):
        fitness=fitness+[FitnessFunc(X[i])]    

    Global_Optimum=0                            #Holds the value of C and gamma which gives the best accuracy yet
    Global_Optimum_fitness=float('inf')         #Holds the best value of accuracy yet
    iteration=[]                                #Holds the number of iterations in a list (to make the convergence graph)
    All_Global_optimums=[]                      #Holds all values of all accuracies which were once a global optimum (convergence graph)
    All_optimums=[]                             #Holds the value of the best of accuracy (from the current population) for every iteration

    t=1                                        #Iterator


    while(t<=T):
        iteration.append(t)
        print("X for iteration=",t)


        for i in range(N):                      #Bringing the population in range                 
            BringRange(X[i])


        Xm=np.average(X,axis=0)                 #Finding the average


        Xrabbit_fitness=fitness[0]              #Finding the rabbit (the hawk that give the maximum accuracy)
        Xrabbit=X[0]
        for i in range(N):
            if fitness[i]>Xrabbit_fitness:
                Xrabbit_fitness=fitness[i]
                Xrabbit=X[i]

        r1=random.random()
        r2=random.random()
        r3=random.random()
        r4=random.random()
        r5=random.random()

        for i in range(0,N):                    #FOR EVERY HAWK
            E0 = random.uniform(-1, 1)
            J = 2 * (1 - random.random())
            E = 2 * E0 * (1 - t/T)



            if abs(E)>=1:                                            #Exploration
                Xrand_ind = random.randint(0,N-1)
                Xrand = X[Xrand_ind]
                q=random.random()                                    #Probability of perching strategy
                if(q>=0.5):                                          #First perching strategy
                    X[i] = Xrand - np.multiply(X[i],2*r2)
                    X[i] = np.multiply(abs(X[i]),r1)
                    X[i] = Xrand-X[i]
                    BringRange(X[i])
                else:                                                #Second perching strategy
                    X[i] = r3 *(lb+r4*(ub-lb))
                    X[i] = Xrabbit - Xm - X[i]
                    BringRange(X[i])



            else:                                                   #Exploitation
                r=random.random()
                if r>=0.5 and abs(E)>=0.5:                          #Soft Besiege
                    deltaX = Xrabbit - X[i]
                    X[i] = deltaX - E * abs(J*Xrabbit - X[i])
                    BringRange(X[i])

                elif r>=0.5 and abs(E)<0.5:                         #Hard Besiege
                    deltaX = Xrabbit - X[i]
                    X[i] = Xrabbit - E*abs(deltaX)
                    BringRange(X[i])

                elif r<0.5 and abs(E)>=0.5:                         #Soft Besiege with rapid dives
                    Y = Xrabbit - E*abs(J*Xrabbit - X[i])
                    S = np.random.rand(dim)
                    Z = Y+S*LF()
                    BringRange(Y)
                    BringRange(Z)
                    if FitnessFunc(Y)<fitness[i]:
                        X[i]=Y
                    if FitnessFunc(Z)<fitness[i]:
                        X[i]=Z

                elif r<0.5 and abs(E)<0.5:                          #Hard Besiege with rapid dives
                    Y = Xrabbit - E*abs(J*Xrabbit - Xm)
                    S = np.random.rand(dim)
                    Z = Y+S*LF()  
                    BringRange(Y)
                    BringRange(Z)
                    if FitnessFunc(Y)<fitness[i]:
                        X[i]=Y
                    if FitnessFunc(Z)<fitness[i]:
                        X[i]=Z


        fitness=[]                                                  #Calculating Fitness for each Hawk  
        for i in range(N):
            fitness=fitness+[FitnessFunc(X[i])]


        optimal_fitness=fitness[0]                                 #Finding the best hawk for this iteration
        optimal=X[0]
        for i in range(N):
            if fitness[i]>optimal_fitness:
                optimal_fitness=fitness[i]
                optimal=X[i]

        All_optimums.append(optimal_fitness)


        if (optimal_fitness>Global_Optimum_fitness or t==1):     #Seeing if the optimum for this iteration is better than the global optimum up until now
            Global_Optimum_fitness=optimal_fitness
            Global_Optimum=optimal


        All_Global_optimums.append(Global_Optimum_fitness)      #Stores every value which was once a global optimum

        t=t+1

    fig1=plt.figure(r+1)
    plt.plot(iteration, All_Global_optimums)
    plt.xlabel("Iteration")
    plt.ylabel("Global Optimum Accuracy")
    plt.title("Convergence Graph")
    plt.grid()
    fig1.show()
 

    
    print()
    print("For ",T, " interations and ",N, " population")
    print("Optimal Parameters=",Global_Optimum)
    print("Optimal fitness=",Global_Optimum_fitness)
    result.append(Global_Optimum_fitness)


```
